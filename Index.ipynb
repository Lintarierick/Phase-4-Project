```python
import pandas as pd
#load datasets

# List of dataset files
files = ['movies.csv', 'links.csv', 'ratings.csv', 'tags.csv']

# Dictionary to store DataFrames
dfs = {}

# Load each dataset and store it in the dictionary
for file in files:
    df_name = file.split('.')[0]  # Extract DataFrame name from the file name
    dfs[df_name] = pd.read_csv(file)

# Display information about each DataFrame
for df_name, df in dfs.items():
    print(f"\nInfo for {df_name}:")
    print(df.info())

    # print the first few rows of the DataFrame
    print(f"\nFirst few rows of {df_name}:")
    print(df.head())
```

```python
# Merging datasets using movieId as the key
merged_df = pd.merge(dfs['movies'], dfs['links'], on='movieId', how='left')
merged_df = pd.merge(merged_df, dfs['ratings'], on='movieId', how='left')
merged_df = pd.merge(merged_df, dfs['tags'], on='movieId', how='left')

# Display information about the merged DataFrame
print("\nInfo for merged DataFrame:")
print(merged_df.info())

# Display summary statistics for the merged DataFrame
print("\nSummary statistics for merged DataFrame:")
print(merged_df.describe())
```
